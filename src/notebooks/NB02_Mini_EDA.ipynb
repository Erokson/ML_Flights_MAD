```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import bootcampviztools as bct

df_mad = pd.read_csv('..\data_sample\mad_clean.csv')
```

```python
bct.cardinalidad(df_mad)
```

| Nombre del campo | Descripción                                | Tipo de variable     | Importancia Inicial | Notas                                                       |
|------------------|--------------------------------------------|----------------------|---------------------|-------------------------------------------------------------|
| Fecha            | Fecha del registro del tráfico             | Numérica Discreta    | 2                   | Puede usarse como referencia temporal para análisis mensuales |
| Pasajeros        | Número total de pasajeros                  | Numérica Continua    | 3                   | Variable objetivo en muchos análisis de tráfico              |
| Origen           | Aeropuerto de origen                       | Numérica Discreta    | 2                   | Útil para segmentar por aeropuerto de salida                |
| Destino          | Aeropuerto de destino                      | Numérica Discreta    | 2                   | Permite análisis por punto de llegada                       |
| Tipo de tráfico  | Clasificación entre Llegadas y Salidas     | Binaria              | 2                   | Permite distinguir el tipo de movimiento                    |
| Año              | Año del registro                           | Categórica           | 1                   | Útil para análisis estacionales o evolutivos                |
| Mes              | Mes del registro                           | Categórica           | 1                   | Clave para análisis de estacionalidad mensual               |

```python
# Gráfico de evolución global

# Definir pasajeros_mensual si no está definido
pasajeros_mensual = df_mad.groupby("Fecha")["Pasajeros"].sum()

plt.figure(figsize=(14, 6))
pasajeros_mensual.plot()
plt.title("Evolución mensual del total de pasajeros")
plt.ylabel("Pasajeros")
plt.xlabel("Fecha")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Guardamos el gráfico
plt.savefig('..\img\mad_evolucion_global.png')

# Gráfico de evolución por tipo de tráfico
mensual_tipo = df_mad.groupby(['Fecha', 'Tipo de trafico'])['Pasajeros'].sum().unstack()
mensual_tipo.plot(figsize=(14, 6))
plt.title("Evolución mensual de pasajeros por tipo de tráfico")
plt.ylabel("Pasajeros")
plt.xlabel("Fecha")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#Guardamos el gráfico
plt.savefig('..\img\mad_evolucion_tipo.png')

# Grafico de estacionalidad año a año 
fig, axes = plt.subplots(1, 1, figsize=(10, 10), sharex=True)
for year in pivot_heatmap.index:
    axes.plot(pivot_heatmap.columns, pivot_heatmap.loc[year].values, label=year)
plt.legend()
plt.ylabel("Pasajeros")
plt.xticks(ticks=range(0, 12), labels=["Ene", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"])
plt.xlabel("Mes")
plt.title("Evolución mensual de pasajeros por año")
plt.show()
#Guardamos el gráfico 
plt.savefig('..\img\mad_evolucion_anual.png')
```

```python
# Gráfico de calor: pasajeros por mes y año

pivot_heatmap = df_mad.pivot_table(index="Año", columns="Mes", values="Pasajeros", aggfunc="sum")
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_heatmap, cmap="YlGnBu", annot=True, fmt=".0f", linewidths=.5)
plt.title("Heatmap de pasajeros por año y mes")
plt.xlabel("Mes")
plt.ylabel("Año")
plt.tight_layout()
plt.show()
#Guardamos el gráfico
plt.savefig('..\img\mad_heatmap.png')

# Calcular la media móvil de 12 meses sobre el total de pasajeros mensual
pasajeros_mensual_rolling = pasajeros_mensual.rolling(12).mean()
pasajeros_mensual_rolling.plot(label="Tendencia (Media móvil 12 meses)")
plt.xlabel("Fecha")
plt.ylabel("Media Pasajeros (Ventana 12 meses)")
plt.legend()
plt.show()
#Guardamos el gráfico
plt.savefig('..\img\mad_media_movil.png')
```

 **Distribución anual y mensual (Heatmap)**:
   - El año 2020 destaca por ser el más afectado, con mínimos de tráfico en todos los meses.
   - La recuperación es evidente a partir de 2021, con especial incremento en los meses estivales de los años 2022 a 2024.
   - 2024 presenta los valores más altos registrados en toda la serie.
